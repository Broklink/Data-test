```python
def custom_summary(df):
    numeric_df = df.select_dtypes(include=[np.number])
    summary = numeric_df.apply(lambda x: pd.Series({
        'mean': x.mean(),
        'std': x.std(),
        'min': x.min(),
        '25%': x.quantile(0.25),
        'median': x.median(),
        '75%': x.quantile(0.75),
        'max': x.max(),
        'skew': x.skew(),
        'kurtosis': x.kurt()
    }))
    return summary.round(2)

custom_summary(df)
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Rating'], bins=10, kde=True, color='purple')
plt.title('Distribution of Player Rating')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

# Comparison â€“ à¸à¸²à¸£à¹€à¸›à¸£à¸µà¸¢à¸šà¹€à¸—à¸µà¸¢à¸šà¸£à¸°à¸«à¸§à¹ˆà¸²à¸‡à¸à¸¥à¸¸à¹ˆà¸¡

```python
top_clutch = df.sort_values(by='Clutch_Attempted', ascending=False).head(10)
top_clutch.plot(kind='bar', x='Player', y=['Clutch_Won', 'Clutch_Attempted'], figsize=(10,6))
plt.title('Clutch Won vs Attempted (Top 10 Players)')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
